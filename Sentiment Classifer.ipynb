```python
sent=['neg','mid','pos']
data=[]
for i in range(len(sent)):
    with open('comments_sets/'+sent[i]+' corpus.txt','r',encoding='utf-8') as f:
        reviews = f.readlines()
        for r in reviews:
            data.append([i-1,r[:-2]])
```

```python
import pandas as pd
df=pd.DataFrame(data)
df
```

```python
df=df[~df[1].isin([''])]
```

```python
df
```

```python
df.columns=['sentiment','comment']
```

```python
print(df.shape)
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer  # Convert a collection of text documents to a matrix of token counts
vectorizer = CountVectorizer()  # In scikit-learn, everything with a fit_transform is an instance of some type, which is to say that you'll need to initialize that instance first, where you are calling fit_transform as if it were a staticmethod.
bow = vectorizer.fit_transform(df.comment)
```

```python
bow.shape
```

```python
from sklearn.model_selection import train_test_split  
x_train,x_test,y_train,y_test = train_test_split(bow,df.sentiment,random_state=1)  
print('è®­ç»ƒé›†ï¼š'+str(x_train.shape)+' '+str(y_train.shape))
print('æµ‹è¯•é›†ï¼š'+str(x_test.shape)+' '+str(y_test.shape))
```

```python
# ä½¿ç”¨è´å¶æ–¯é¢„æµ‹åˆ†ç±»
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

#æ ¹æ®æµ‹è¯•é›†çš„bowç‰¹å¾ï¼Œè®­ç»ƒè´å¶æ–¯åˆ†ç±»å™¨
clf = nb.fit(x_train,y_train)

#åŸºäºæµ‹è¯•é›†çš„bowç‰¹å¾è¿›è¡Œé¢„æµ‹
y_pred = clf.predict(x_test) 

#pythonæµ‹é‡å·¥å…·é›†
from sklearn import metrics 

#å‡†ç¡®ç‡æµ‹è¯•
accuracy=metrics.accuracy_score(y_test,y_pred)
print('å‡†ç¡®ç‡ï¼š'+str(accuracy))

#æ··æ·†çŸ©é˜µ
print('æ··æ·†çŸ©é˜µï¼š\n'+str(metrics.confusion_matrix(y_test,y_pred)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
clf = knn.fit(x_train, y_train)
y_pred = clf.predict(x_test)
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, y_pred)
print('å‡†ç¡®ç‡ï¼š' + str(accuracy))
print('æ··æ·†çŸ©é˜µï¼š\n' + str(metrics.confusion_matrix(y_test, y_pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
clf = rf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, y_pred)
print('å‡†ç¡®ç‡ï¼š' + str(accuracy))
print('æ··æ·†çŸ©é˜µï¼š\n' + str(metrics.confusion_matrix(y_test, y_pred)))
```
